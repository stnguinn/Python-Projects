```python
import pandas as pd

# Load the CSV file from the URL
url = 'https://gist.githubusercontent.com/mattkram/9684863843254402942dfede27af2cb7/raw/2590dd8185b833aacf247c0595edbb07a025a6d7/Smithsonian_VOTW_Holocene_Volcanoes.csv'
volcanoes_df = pd.read_csv(url)

# Display summary statistics for numerical columns
print("Summary statistics for numerical columns:")
print(volcanoes_df.describe())

# Display summary statistics for all columns, including categorical
print("\nSummary statistics for all columns (including categorical):")
print(volcanoes_df.describe(include='all'))
```

```python
#explore missing values
print("\nMissing values in each column:")
print(volcanoes_df.isnull().sum())
```

```python
#Frequency Counts for Key Categorical Columns
categorical_cols = volcanoes_df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\nUnique values in '{col}':")
    print(volcanoes_df[col].unique())
```

```python
#Summary Statistics by Group
# check the actual column names
print("Column names in the DataFrame:", volcanoes_df.columns.tolist())
```

```python
# Summary Statistics by Group
print("\nAverage elevation by volcano type:")
print(
    volcanoes_df.groupby('Primary_Volcano_Type')['Elevation']
    .mean()
    .sort_values(ascending=False)
    .round(2)
)
```

```python
#Correlation Matrix for Numerical Columns
import seaborn as sns
import matplotlib.pyplot as plt

print("\nCorrelation matrix:")
correlation = volcanoes_df.corr(numeric_only=True)
print(correlation)

# Plot
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Histogram of Elevation distribution
plt.figure(figsize=(8, 5))
sns.histplot(volcanoes_df['Elevation'].dropna(), bins=30, kde=True)
plt.title('Distribution of Volcano Elevation')
plt.xlabel('Elevation')
plt.ylabel('Frequency')
plt.show()
```

```python
# World Map Plot of Volcano Locations
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=volcanoes_df,
    x='Longitude',
    y='Latitude',
    hue='Primary_Volcano_Type',
    alpha=0.7,
    edgecolor=None,
    legend=False
)
plt.title('Geographic Distribution of Volcanoes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
#  Volcano Type Risk Categories
# Add a rough risk level based on volcano type
risk_map = {
    'Stratovolcano': 'High',
    'Caldera': 'High',
    'Complex': 'Medium',
    'Shield': 'Low',
    'Lava dome': 'Medium',
    'Pyroclastic cone': 'Medium',
    'Submarine': 'Low',
    'Unknown': 'Unknown',
}

volcanoes_df['Risk Level'] = volcanoes_df['Primary_Volcano_Type'].map(risk_map)

# Show count per risk level
print("Volcano counts by assigned risk level:")
print(volcanoes_df['Risk Level'].value_counts())
```

```python
#Elevation and Risk Consideration
# Show average elevation by risk level
print("\nAverage elevation by risk level:")
print(
    volcanoes_df.groupby('Risk Level')['Elevation']
    .mean()
    .round(2)
)
```

```python
#Country Risk: Volcano Count by Country
print("\nTop 10 countries with the most volcanoes (potential hotspots):")
print(volcanoes_df['Country'].value_counts().head(10))
```

```python
#Recent Activity Indicator
# Filter recent eruptions (example: after 1900)
recent_eruption = volcanoes_df[
    volcanoes_df['Last_Eruption_Year']
    .astype(str)
    .str.contains('20|19', na=False)
]

print("\nVolcanoes with eruptions in the 20th or 21st century:")
print(recent_eruption[['Volcano_Name', 'Country', 'Last_Eruption_Year']].head(10))
```

```python
#High-Risk Volcanoes Summary
# Flag volcanoes as high-risk if they are stratovolcanoes or calderas, have erupted recently, and are over 2000m
# Define 'Erupted Recently' as True if eruption year is 1900 or later
volcanoes_df['Erupted Recently'] = volcanoes_df['Last_Eruption_Year'] >= 1900

# Filter for high-risk volcanoes
high_risk = volcanoes_df[
    (volcanoes_df['Risk Level'] == 'High') &
    (volcanoes_df['Erupted Recently']) &
    (volcanoes_df['Elevation'] > 2000)
]

# Display result
print("\nHigh-risk volcanoes (Type = High Risk, Recent Eruption, Elevation > 2000m):")
print(high_risk[['Volcano_Name', 'Country', 'Primary_Volcano_Type', 'Elevation', 'Last_Eruption_Year']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load dataset
url = 'https://gist.githubusercontent.com/mattkram/9684863843254402942dfede27af2cb7/raw/2590dd8185b833aacf247c0595edbb07a025a6d7/Smithsonian_VOTW_Holocene_Volcanoes.csv'
volcanoes_df = pd.read_csv(url)

# Define basic risk levels
risk_map = {
    'Stratovolcano': 'High',
    'Caldera': 'High',
    'Complex': 'Medium',
    'Shield': 'Low',
    'Lava dome': 'Medium',
    'Pyroclastic cone': 'Medium',
    'Submarine': 'Low',
    'Unknown': 'Unknown',
}
volcanoes_df['Risk Level'] = volcanoes_df['Primary_Volcano_Type'].map(risk_map)

# Drop rows with missing coordinates
volcanoes_df = volcanoes_df.dropna(subset=['Latitude', 'Longitude'])

# Set up flat world map projection
plt.figure(figsize=(16, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()

# Add features
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.set_title('Global Volcano Risk Levels', fontsize=16)

# Plot volcanoes by risk level
risk_colors = {'High': 'red', 'Medium': 'orange', 'Low': 'green', 'Unknown': 'gray'}

for risk, color in risk_colors.items():
    subset = volcanoes_df[volcanoes_df['Risk Level'] == risk]
    ax.scatter(
        subset['Longitude'],
        subset['Latitude'],
        color=color,
        s=10,
        label=risk,
        alpha=0.6,
        transform=ccrs.PlateCarree()
    )

plt.legend(title='Risk Level', loc='lower left')
plt.show()
```
